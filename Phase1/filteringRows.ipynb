# Filtering rows based on conditions

```python
import pandas as pd
```

```python
data = {
    'Name' : ['ram','shayam','rohit','virat','aditi','raj','driti','simran'],
    'Age' : [20,23,34,26,21,34,29,31],
    'Salary' : [50000,60000,76000,65000,64000,90000,56000,87000],
    'Performance Score' : [85,90,95,94,89,29,99,100]
}
```

```python
df = pd.DataFrame(data)

# Simple data frame
print('Simple Dataframe :\n')
print(df)
```

```python
# Filtering rows - single condition

high_salary = df[df['Salary'] > 50000]
print('People with salary > 50k:\n')
print(high_salary)
```

```python
# Filtering rows - multiple condition

salary = df[(df['Salary'] > 50000) & (df['Salary'] < 90000)]
print('People with salary > 50k and salary < 90k:\n')
print(salary)
```
