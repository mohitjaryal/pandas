# Merging, Joining and Concatenating

```python
import numpy as np
import pandas as pd
```

```python
# importing data sets
deliveries = pd.read_csv('data/deliveries.csv')
matches = pd.read_csv('data/matches.csv')
nov = pd.read_csv('data/reg-month1.csv')
dec = pd.read_csv('data/reg-month2.csv')
students = pd.read_csv('data/students.csv')
courses = pd.read_csv('data/courses.csv')
```

```python
# pd.concat -> data frame will be vertically merged (make sure that the data frames have same number of columns)
regs = pd.concat([nov,dec],ignore_index=True)
regs
```

```python
# MultiIndex Data Frame -> means more than 2 index 
multi_idx = pd.concat([nov,dec],keys=['Nov','Dec'])
multi_idx
```

```python
 # Accessing only Nov index 
multi_idx.loc['Nov']
```

```python
# Accessing the 3rd index of Nov
multi_idx.loc[('Nov',3)]
```

```python
# Accessing the 6rd index of dec
multi_idx.loc[('Dec',6)]
```

```python
# Horizontally stacking two df
pd.concat([nov,dec],axis=1)
```
