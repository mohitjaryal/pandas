# Selecting Specific column, filtering rows, combine multiple conditions

```python
import pandas as pd
```

```python
data = {
    'Name' : ['ram','shayam','rohit','virat','aditi','raj','driti','simran'],
    'Age' : [20,23,34,26,21,34,29,31],
    'Salary' : [50000,60000,76000,65000,64000,90000,56000,87000],
    'Performance Score' : [85,90,95,94,89,29,99,100]
}
```

```python
df = pd.DataFrame(data)

# Simple data frame
print('Simple Dataframe :\n')
print(df)
```

```python
# Accesing single column
print('Accessing Single column : \n')
print(df['Name'])
```

```python
# Accesing multiple column
print('Accessing multiple columns:\n')
print(df[['Name', 'Salary']])
```
