## Long VS Wide data

#### Wide format is where we have a single row for every data point with multiple columns to hold the values of various attributes. (No. of columns more, rows less)

#### Long format is where, for each data point we have as many rows as the number of attributes and each row contains the value of a particular attribute for a given data point.(No of rows more, columns less)

```python
# melt -> it converts wide data format into long data format
melt_df = pd.DataFrame({
    'cse':[140],
    'cse_ai':[200],
    'cse_ml':[70]
})

# performing melt in melt_df
melt_df.melt(var_name='branch',value_name='no. of students')
```

```python
# next example
pd.DataFrame({
    'branch':['cse','cse_ai','cse_gen_ai','ece'],
    '2023':[100,210,130,213],
    '2024':[190,329,340,489],
    '2025':[290,870,652,832],
    '2026':[500,600,800,900]
} 
).melt(id_vars=['branch'],var_name='year',value_name='students') # iid_vars = columns you want to keep unchanged (as identifiers)
```

#### Melt using real world example (Working with covid data sets)

```python
confirm = pd.read_csv('data/time_series_covid19_confirmed_global.csv')
confirm
```

```python
death = pd.read_csv('data/time_series_covid19_deaths_global.csv')
death
```

```python
# convert these above data sets into a df which have 2 columns -> name,country,confirm cases, death cases

# using melt -> changing wide data into long data
death = death.melt(id_vars=['Province/State','Country/Region','Lat','Long'],var_name='date',value_name='num_deaths')


confirm = confirm.melt(id_vars=['Province/State','Country/Region','Lat','Long'],var_name='date',value_name='num_cases')

# merge
confirm.merge(death, on=['Province/State','Country/Region','Lat','Long','date'])[['Country/Region','date','num_cases','num_deaths']]
```

#### Pivot table 
##### The pivot table takes simple column-wise data as input, and groups the entries into a two-dimensional table that provides a multidimensional summarization of the data.

```python
import seaborn as sns
```

```python
df = pd.read_csv('data/tips.csv')
df
```

```python
# pivot table
df.pivot_table(index='sex',columns='smoker',values='total_bill')
```

```python
# aggr function

# total bill
df.pivot_table(index='sex',columns='smoker',values='total_bill',aggfunc='sum')
```

```python
# multidimensional
df.pivot_table(index=['sex','smoker'],columns=['day','time'],values='total_bill')
```

```python
# margins -> calculate the sum
df.pivot_table(index='sex',columns='smoker',values='total_bill',aggfunc='sum',margins=True)
```
