# Adding calculated data 

```python
import pandas as pd
```

```python
data = {
    'Name' : ['ram',None,'rohit','virat','aditi','raj','driti','simran'],
    'Age' : [20,None,34,26,21,34,29,31],
    'Salary' : [50000,None,60000,65000,64000,90000,56000,87000],
    'Performance Score' : [85,None,95,94,89,29,99,100]
}
```

```python
df = pd.DataFrame(data)
print('Original data :\n',df)
```

# Adding calculated data

```python
# adding mean of the entire Age column 
df['Age'].fillna(df['Age'].mean(),inplace=True) 
print(df)
```

```python
# adding mean of the entire Salary column 
df['Salary'].fillna(df['Salary'].mean(),inplace=True) 
print(df)
```

```python
# adding mean of the entire Performance Score column 
df['Performance Score'].fillna(df['Performance Score'].mean(),inplace=True) 
print(df)
```
