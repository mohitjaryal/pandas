# Handling missing data by adding value to it

```python
import pandas as pd
```

```python
data = {
    'Name' : ['ram',None,'rohit','virat','aditi','raj','driti','simran'],
    'Age' : [20,None,34,26,21,34,29,31],
    'Salary' : [50000,None,60000,65000,64000,90000,56000,87000],
    'Performance Score' : [85,None,95,94,89,29,99,100]
}
```

```python
df = pd.DataFrame(data)
print('Original data :\n',df)
```

```python
df.fillna(0,inplace=True)
print('After adding :\n',df)
```
