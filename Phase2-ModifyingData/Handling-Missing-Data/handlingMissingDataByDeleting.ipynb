# Handling missing data by deleting it

```python
import pandas as pd
```

```python
data = {
    'Name' : ['ram',None,'rohit','virat','aditi','raj','driti','simran'],
    'Age' : [20,None,34,26,21,34,29,31],
    'Salary' : [50000,60000,None,65000,64000,90000,56000,87000],
    'Performance Score' : [85,None,95,94,89,29,99,100]
}
```

```python
df = pd.DataFrame(data)
print('Original data :\n',df)
```

## Two ways to handle missing data - 
### 1. delete it if it is not important  using .dropna()
### 2. fill it with some data 


```python
 
# deleting values using .dropna(axis=0/1,inplace=True/False) - 0 means delete rows and  1 means columns
df.dropna(inplace=True)
print(df)
```
