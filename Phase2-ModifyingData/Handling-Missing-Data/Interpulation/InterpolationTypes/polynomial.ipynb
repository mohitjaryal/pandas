# Polynimial interpolation

```python
import pandas as pd

# sample data 
data = {
    'x': [1, 2, 3, 4, 5, 6],
    'y': [10, None, 30, None, 50, None]
}

df = pd.DataFrame(data)
print("Original Data:\n", df)

# Perform polynomial interpolation (order = 2) - order 2 means Quadratic
df['y'] = df['y'].interpolate(method='polynomial', order=2)

print("\nAfter Polynomial Interpolation:\n", df)
```
