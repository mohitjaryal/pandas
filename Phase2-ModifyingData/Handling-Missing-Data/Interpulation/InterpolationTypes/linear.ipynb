# Linear interpolation

```python
import pandas as pd

data = {
    'values' : [1,2,3,4,5,6],
    'time':[10,None,30,None,50,None]
}

df = pd.DataFrame(data)
print('Original :\n',df)

# performing interpolation in time
df['time'] = df['time'].interpolate(method='linear')
print('After interpolation :\n',df)
```
