# Identifying missing data

```python
import pandas as pd
```

```python
data = {
    'Name' : ['ram',None,'rohit','virat','aditi','raj','driti','simran'],
    'Age' : [20,None,34,26,21,34,29,31],
    'Salary' : [50000,60000,None,65000,64000,90000,56000,87000],
    'Performance Score' : [85,None,95,94,89,29,99,100]
}
```

```python
df = pd.DataFrame(data)
print('Original Data :\n',df)
```

```python
# using isnull() method
print(df.isnull())
```

```python
# if we want to know how many values are missing use .sum()

print(df.isnull().sum())
```
