# Removing data

```python
import pandas as pd
```

```python
data = {
    'Name' : ['ram','shayam','rohit','virat','aditi','raj','driti','simran'],
    'Age' : [20,23,34,26,21,34,29,31],
    'Salary' : [50000,60000,76000,65000,64000,90000,56000,87000],
    'Performance Score' : [85,90,95,94,89,29,99,100]
}
```

```python
df = pd.DataFrame(data)
print('Original data :\n',df)
```

```python
# adding id column using insert method
df.insert(0,"ID",[10,20,30,40,50,60,70,80])
print('After adding ID :\n',df)
```

```python
# removing single column
df.drop(columns=['ID'],inplace=True)
print('Deleting ID Column :\n',df)
```
