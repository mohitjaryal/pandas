# Summary Statistics 
## It provides the numeric summary of column

```python
import pandas as pd
```

```python
data = {
    'name' : ['ram','arun','shayam'],
    'age' : [10,12,19],
    'id' : [1,2,3 ]
}
```

```python
# making data frame
df = pd.DataFrame(data)
print('Oroginal :\n',df)
```

## Summary statistics


```python
# mean()
avg1 = df['age'].mean()
print('Mean (age):\n',avg1)

avg2 =df['id'].mean()
print('Mean (id):\n',avg2)
```

```python
# total value .sum()

sum1 = df['age'].sum()
print('Sum (age):\n',sum1)

sum2 =df['id'].sum()
print('Sum (id):\n',sum2)
```

```python
# min value .min()

min1 = df['age'].min()
print('Min (age):\n',min1)

min2 =df['id'].min()
print('Min (id):\n',min2)
```

```python
# max value .max()

max1 = df['age'].max()
print('Max (age):\n',max1)

max2 =df['id'].max()
print('Max (id):\n',max2)
```
