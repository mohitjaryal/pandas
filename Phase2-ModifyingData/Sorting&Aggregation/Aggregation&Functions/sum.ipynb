# sum.() aggregation function

```python
import pandas as pd

data = {
    'Name': ['Amit', 'Riya', 'John', 'Sara', 'Tom'],
    'Age': [25, 30, 28, 22, 35],
    'Salary': [50000, 60000, 58000, 52000, 70000]
}

df = pd.DataFrame(data)
print("Original Data:\n", df)

# .sum() Aggregation Function
print("\nSum of each column:\n", df[['Age', 'Salary']].sum())
```
