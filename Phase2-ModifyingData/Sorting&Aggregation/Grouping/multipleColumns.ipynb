# Grouping multiple columns

```python
import pandas as pd
```

```python
data = {
    'Name' : ['ram','shayam','rohit','virat','aditi','raj','driti','simran'],
    'Age' : [20,23,33,20,21,30,23,31],
    'Salary' : [50000,60000,76000,65000,64000,90000,56000,87000],
    'Performance Score' : [85,90,95,94,89,29,99,100]
}
```

```python
df = pd.DataFrame(data)
print('Original :\n',df)
```

```python
# we'll group name & age and then calculate their salary
grouped = df.groupby(['Name','Age'])['Salary'].sum()
print('After grouping :\n',grouped)
```
