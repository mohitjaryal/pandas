# Updating values

```python
import pandas as pd
```

```python
data = {
    'Name' : ['ram','shayam','rohit','virat','aditi','raj','driti','simran'],
    'Age' : [20,23,34,26,21,34,29,31],
    'Salary' : [50000,60000,76000,65000,64000,90000,56000,87000],
    'Performance Score' : [85,90,95,94,89,29,99,100]
}
```

```python
df = pd.DataFrame(data)
print('Original :\n',df)
```

```python
# Updating values
df.loc[6,'Name'] = 'driti sharma'
print('After updating value :\n',df)
```

```python
# Updating values 
df.loc[6,'Age'] = 30
print('After updating value :\n',df)
```

```python
# incresing salary by 5% - more than one value (entire row)

# updating value
df['Salary'] = df['Salary'] * 1.05
print('Incrementing salary by 5% :\n',df)
```

```python
# Updating age by +1 years
df['Age'] = df['Age'] + 1
print('After updating age :\n',df)
```
