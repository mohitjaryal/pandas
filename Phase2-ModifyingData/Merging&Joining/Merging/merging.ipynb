# Merging 

```python
import pandas as pd

# data frame 1 - customers_detail
customers_detail = pd.DataFrame({
    'name' : ['ramesh','suresh','mahesh','rajesh'],
    'year' : [2024,2024,2024,2024],
    'id' : [1,2,3]
})

# data frame 2 - orders
orders = pd.DataFrame({
    'id' : [3,2,9],
    'amount' : [100,200,900]
})

# merging

pd.merge(customers_detail,orders, on='id',how= )
```
