# left merging (joining)

```python
import pandas as pd

# data frame 1 - customers_detail
customers_detail = pd.DataFrame({
    'name' : ['ramesh','suresh','mahesh','rajesh'],
    'year' : [2024,2024,2024,2024],
    'id' : [1,2,3,4]
})

# data frame 2 - orders
orders = pd.DataFrame({
    'id' : [8,9,1,4],
    'amount' : [100,200,900,189]
})

# merging

merged = pd.merge(customers_detail,orders, on='id',how='left')
print('Left joined :\n',merged)
```
