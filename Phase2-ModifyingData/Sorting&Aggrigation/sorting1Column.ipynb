# Sorting data - 1 column

```python
import pandas as pd
```

```python
data = {
    'name' : ['ram','arun','shayam'],
    'age' : [10,12,19],
    'id' : [1,2,3 ]
}
```

```python
# creating data frame
df = pd.DataFrame(data)
print('Original :\n',df)
```

```python
# sorting in descending order
df.sort_values(by='name',ascending=False,inplace=True)
print('After sorting in Descending order :\n',df)
```
