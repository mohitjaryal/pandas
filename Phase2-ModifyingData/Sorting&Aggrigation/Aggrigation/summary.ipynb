# Summary Statistics 
## It provides the numeric summary of column

```python
import pandas as pd
```

```python
data = {
    'name' : ['ram','arun','shayam'],
    'age' : [10,12,19],
    'id' : [1,2,3 ]
}
```

```python
# making data frame
df = pd.DataFrame(data)
print('Oroginal :\n',df)
```

## Summary statistics


```python
# mean()
avg1 = df['age'].mean()
print('Mean (age):\n',avg1)

avg2 =df['id'].mean()
print('Mean (id):\n',avg2)
```

```python
# total value .sum()

sum1 = df['age'].sum()
print('Mean (age):\n',sum1)

sum2 =df['id'].sum()
print('Mean (id):\n',sum2)
```
