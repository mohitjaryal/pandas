# Sorting multiple columns

```python
import pandas as pd
```

```python
data = {
    'name' : ['ram','arun','shayam','bob','rohit'],
    'age' : [10,12,19,20,16],
    'id' : [3,2,1,9,5]
}
```

```python
# creating data frame
df = pd.DataFrame(data)
print('Original :\n',df)
```

```python
# sorting in Ascending order
df.sort_values(by=['age','id'],ascending=True,inplace=True)
print('After sorting in Ascending order :\n',df)
```
