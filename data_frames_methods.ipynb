```python
# value_counts() -> applicable in both series and data frame

# making a data frame
marks = pd.DataFrame([
    [100,54,23],
    [89,64,45],
    [66,75,64],
    [89,64,45],
    [89,64,45],
    [90,100,89],
    [90,100,89],
    [89,64,45],
    [90,100,89]
],columns=['iq','maks','package'])

marks
```

```python
# value_counts() -> applicable in both series and data frame
# value_counts method counts how many rows in a data frame and frequency of the repetation of those rows
marks.value_counts()
```

```python
# find which player has won the most potm (man of the match) in finals and semi finals
ipl[~ipl['MatchNumber'].str.isdigit()]['Player_of_Match'].value_counts()
```
