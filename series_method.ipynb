## Series methods

```python
# astype -> change the data type
import sys
sys.getsizeof(vk)
```

```python
# change data type ( to int16
vk.astype('int16')
```

```python
# clip 
subs.clip(100,200)
```

```python
# drop_duplicates
temp = pd.Series([1,2,3,2,4,5,4,7,6,5])
temp.drop_duplicates()
```

```python
# duplicated -> check is there any duplicated values or not
temp.duplicated().sum()
```

```python
# size -> counts all the values included null

# creating a series 
temp2 = pd.Series([1,2,np.nan,3,2,4,5,np.nan,5,np.nan])
temp2.drop_duplicates()
temp2.size
```

```python
# count -> it dosen't count null values
temp2.count()
```

```python
# isnull -> check how many missing values in data
vk.isnull().sum()
```

```python
# dropna -> remove all missing values
# it creates a new series
temp2.dropna()
```

```python
# fillna -> to fill the missing values
# it creates a new series
temp2.fillna(0)
```

```python
# isin -> check multiple things in data
vk[vk.isin([49,99])]
```

```python
    # apply -> helps to apply custom logic in series
subs['Subscribers gained'].apply(lambda x: 'good day' if x > subs['Subscribers gained'].mean() else 'bad day')
```

```python
# copy 
# copy makes copy of the data so if we perform any operation into it the real data will remain unchanged
# when we use .head() it gives the view of the data and if we do any changes into it - it will affect the data
vk.head(2).copy()
```
