## Pandas

### Pandas Series -> it is like a column in a table. It is a 1D array holding data of any type

```python
# importing pandas

import pandas as pd
import numpy as np
```

#### Series from lists

```python
# string 

country =['India','Russia','USA','Japan']

# makeing series using .series() in pandas
# series has twwo things 1. index, 2. value
pd.Series(country)
```

```python
# integers

runs = [1,22,30,100,89,99,56,45,34,43]

# creating series
runs = pd.Series(runs)
runs
```

#### custom indexing in series

```python
marks = [90,23,57,34,23]
subject = ['Math','Hindi','English','Science','CS']

# creating series
# now make marks as value and subjects as index
# using index=(jise index banna cahte ho uska name)
marks = pd.Series(marks,index=subject)
marks
```

#### Setting a name for series objects

```python
# giving a name to this series
# using name parameter
pd.Series(marks,index=subject,name="Rohit's marks")
```

### Series from dictionary

```python
# creating series from dictionary

# dictionary
sub_marks = {
    'math' : 100,
    'english' : 90,
    'science' : 100,
    'history' : 84,
}
```

```python
# creating series from dictionary
marks_series = pd.Series(sub_marks,name="Mohit's marks")
marks_series
```

#### Series Attribute

```python
# size attribute
# this tells us how many items are there in our series

# this will tell us about the size of our series
marks_series.size
```

```python
# dtype
# this will tell us the type of data stored in series

marks_series.dtype
```

```python
# name
# this will tell the name of our siries

marks_series.name
```

```python
# is_unique
# this tells that all the items in a series are unique or not 
marks_series.is_unique
```

```python
# index
# this will display all the index value in series

marks_series.index
```

```python
# values
# this gives index object
# this will display all the value present in series

marks_series.values
```
