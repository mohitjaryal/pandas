## Pandas

### Pandas Series -> it is like a column in a table. It is a 1D array holding data of any type

```python
# importing pandas

import pandas as pd
import numpy as np
```

#### Series from lists

```python
# string 

country =['India','Russia','USA','Japan']

# makeing series using .series() in pandas
# series has twwo things 1. index, 2. value
pd.Series(country)
```

```python
# integers

runs = [1,22,30,100,89,99,56,45,34,43]

# creating series
runs = pd.Series(runs)
runs
```

#### custom indexing in series

```python
marks = [90,23,57,34,23]
subject = ['Math','Hindi','English','Science','CS']

# creating series
# now make marks as value and subjects as index
# using index=(jise index banna cahte ho uska name)
marks = pd.Series(marks,index=subject)
marks
```

#### Setting a name for series objects

```python
# giving a name to this series
# using name parameter
pd.Series(marks,index=subject,name="Rohit's marks")
```

## Excercise

```python
# importing ipl data set
deliveries  = pd.read_csv('data/deliveries.csv')
deliveries.head(2)
```

```python
# Find top 10 batsman in terms of runs
deliveries.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# find the batsman with max numbers of six
six = deliveries[deliveries['batsman_runs'] == 6]
six.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1)
```

```python
# find top 10 batsman with max number of 4's and 6's in the last 5 overs

# filter last 5 overs
last_overs = deliveries[deliveries['over'] > 15]

# keep only 4s and 6s
last_overs = last_overs[(last_overs['batsman_runs'] == 4) | (last_overs['batsman_runs'] == 6)]

# count per batsman
result = last_overs.groupby('batsman')['batsman'].count().sort_values(ascending=False)

print(result.head(10))
```

```python
# find virat kohli's record against all teams

# filtering only where the batsman is virat kohli
temp_df = deliveries[deliveries['batsman'] == 'V Kohli']

temp_df.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
# create a function that can return the highest score of any batsman

def highest(batsman):
    vk_df = deliveries[deliveries['batsman'] == batsman]
    return vk_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(2).values[0]

highest('R Sharma')
```
