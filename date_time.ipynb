### DateTimeIndex object
#### It is a collection of pandas timestamp

```python
# from string 
pd.DatetimeIndex(['2022/1/1','2023/1/1','2024/1/1','2025/1/1','2026/1/1'])
```

```python
# creating datetimeindex using python's datetime
pd.DatetimeIndex([dt.datetime(2024,1,1),dt.datetime(2025,2,2),dt.datetime(2026,5,1)])
```

```python
# using timestamps
dt_index = pd.DatetimeIndex([pd.Timestamp(2024,1,1),pd.Timestamp(2025,2,2),pd.Timestamp(2026,5,1)])
dt_index
```

```python
# using datetimeindex as series index
pd.Series([1,2,3],index=dt_index)
```

#### date range funciton

```python
# generate daily dates in a given range
pd.date_range(start='2026/5/1',end='2026/6/1')
```

```python
# alternate days

# using freq
pd.date_range(start='2026/5/1',end='2026/6/1',freq='2D')
```

```python
# find only Business days (Mon-Fri) -> B
pd.date_range(start='2026/5/1',end='2026/6/1',freq='B')
```

```python
# find one day per week -> W
pd.date_range(start='2026/5/1',end='2026/6/1', freq='W')
```

```python
# find one day per week (Mon) -> W
pd.date_range(start='2026/5/1',end='2026/6/1', freq='W-Mon')
```

```python
# Hourly time stamp

pd.date_range(start='2026/5/1',end='2026/6/1', freq='h')
```

```python
# Hourly time stamp -> every 5 hours

pd.date_range(start='2026/5/1',end='2026/6/1', freq='5h')
```

```python
# Month end

pd.date_range(start='2026/5/1',end='2026/6/1', freq='ME')
```

```python
# Month start

pd.date_range(start='2026/5/1',end='2026/6/1', freq='MS')
```

```python
# Year end -> YE

pd.date_range(start='2026/5/1',end='2030/6/1', freq='YE')
```

```python
# Year start
pd.date_range(start='2026/5/1',end='2030/6/1', freq='YS')
```

```python
# using periods -> 25 days
pd.date_range(start='2026/5/1',periods=25,freq='D')
```

```python
# using periods -> 6hours
pd.date_range(start='2026/5/1',periods=25,freq='6h')
```

#### to_datetime function
##### it converts an existing objects to pandas timestamp/datetime object

```python
# series example

# series
s = pd.Series(['2026/1/1','2025/2/2','2024/3/2','2023/3/1'])

# converting it into a date time
pd.to_datetime(s)
```

```python
# fetch month form s 
pd.to_datetime(s).dt.month
```

```python
# working with errors
s = pd.Series(['2026/1/1','2025/2/2','2024/30/2','2023/3/1'])

# error handing
pd.to_datetime(s, errors='coerce')
```
